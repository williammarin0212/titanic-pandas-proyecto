```python
# 2. Describa las variables que se involucran en los datasets

import pandas as pd

# Cargar los datasets
train_df = pd.read_csv('../data/train.csv')
test_df = pd.read_csv('../data/test.csv')

# 1. Mostrar información general de cada dataset
print("=== Información del dataset de entrenamiento ===")
display(train_df.info())
print("\n=== Información del dataset de prueba ===")
display(test_df.info())

# 2. Descripción estadística de variables numéricas
print("\n=== Descripción estadística del dataset de entrenamiento ===")
display(train_df.describe())

print("\n=== Descripción estadística del dataset de prueba ===")
display(test_df.describe())

# 3. Valores únicos de variables categóricas
categorical_cols = ['Survived', 'Pclass', 'Sex', 'Embarked']
for col in categorical_cols:
    if col in train_df.columns:
        print(f"\nValores únicos de '{col}': {train_df[col].unique()}")
```
