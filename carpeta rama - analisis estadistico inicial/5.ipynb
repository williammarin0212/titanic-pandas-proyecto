```python
import pandas as pd
import os

# Ruta correcta para tu estructura de carpetas
train_df = pd.read_csv(os.path.join("..", "data", "train.csv"))
test_df  = pd.read_csv(os.path.join("..", "data", "test.csv"))

print("Datos cargados correctamente.")
print("Filas train:", len(train_df))
print("Filas test:", len(test_df))
```

```python
# 5. Junte verticalmente los dataframes train y test en un solo conjunto de datos. Antes de hacerlo, agregue una nueva columna llamadasource que indique el origen de cada fila ("train" o "test").

import pandas as pd
import numpy as np

# Agregar columna 'source' para indicar el origen
train_df['source'] = 'train'
test_df['source'] = 'test'

# Crear columna 'Survived' en test con valores NaN
test_df['Survived'] = np.nan

# Reordenar columnas de test para que coincidan con train (opcional)
test_df = test_df[train_df.columns]

# Concatenar verticalmente
combined_df = pd.concat([train_df, test_df], ignore_index=True)

# Revisar las primeras filas
display(combined_df.head())
display(combined_df.tail())
```
