```python
# 11. ¿Cu´al es el la proporci´on de personas del mismo sexo, por cada grupo etario que ten´ıan y no ten´ıan cabina, que sobrevivieron y no? Finalice con una adecuada conclusi´on (an´alisis). Antes de responder, defina una nueva variable llamada grupo etario a partir de la variable Age.
# Usando las tres categor´ıas, por ejemplo:
# • Ni˜nos (menores de 10 a˜nos)
# • Adultos (de 18 a 49 a˜nos)
# • Mayores (50 a˜nos o m´as)

import pandas as pd

# Cargar datos
train = pd.read_csv('../data/train.csv')

# === 1. Crear grupo etario ===
def grupo_etario(edad):
    if pd.isna(edad):
        return None
    elif edad < 10:
        return "Niños"
    elif 18 <= edad <= 49:
        return "Adultos"
    elif edad >= 50:
        return "Mayores"
    else:
        return None

train["Grupo_Etario"] = train["Age"].apply(grupo_etario)

# === 2. Crear variable Cabin_binaria (tiene cabina / no tiene cabina) ===
train["Tiene_Cabina"] = train["Cabin"].notna().map({True: "Con cabina", False: "Sin cabina"})

# === 3. Tabla con proporciones por sexo, grupo etario, cabina y supervivencia ===
tabla = (
    train
    .groupby(["Sex", "Grupo_Etario", "Tiene_Cabina", "Survived"])
    .size()
    .groupby(level=[0,1,2])  # normalizar dentro de cada subgrupo
    .apply(lambda x: x / x.sum()))

print(tabla)
```
