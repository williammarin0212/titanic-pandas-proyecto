```python
# 8. Responda las siguientes preguntas bas´andose en los datos:
# (a) ¿Cu´al es la edad promedio de los pasajeros del Titanic?
# (b) ¿Cu´antos pasajeros sobrevivieron y cu´antos murieron?
# (c) ¿Cu´al es la tarifa promedio pagada por los pasajeros de primeraclase?
# (d) ¿Cu´antos pasajeros viajaron con un familiar a bordo?
# (e) ¿Cu´al es la edad m´as joven y la m´as vieja de los pasajeros?
# (f) ¿Cu´antos pasajeros viajaron desde cada puerto de embarque?
# (g) ¿Cu´antos pasajeros viajaron solos y cu´antos con familiares?

import pandas as pd

# Cargar el dataset
df = pd.read_csv('../data/train.csv')

# Crear variable de familiares (SibSp + Parch)
df["Familiares"] = df["SibSp"] + df["Parch"]

print("\n(a) ¿Cuál es la edad promedio de los pasajeros?")
print(df["Age"].mean())

print("\n(b) ¿Cuántos pasajeros sobrevivieron y cuántos murieron?")
display(df["Survived"].value_counts())

print("\n(c) ¿Cuál es la tarifa promedio pagada por los pasajeros de primera clase?")
print(df[df["Pclass"] == 1]["Fare"].mean())

print("\n(d) ¿Cuántos pasajeros viajaron con al menos un familiar a bordo?")
print((df["Familiares"] > 0).sum())

print("\n(e) ¿Cuál es la edad más joven y la más vieja de los pasajeros?")
print("Edad mínima:", df["Age"].min())
print("Edad máxima:", df["Age"].max())

print("\n(f) ¿Cuántos pasajeros viajaron desde cada puerto de embarque?")
display(df["Embarked"].value_counts())

print("\n(g) ¿Cuántos pasajeros viajaron solos y cuántos con familiares?")
viajaron_solos = (df["Familiares"] == 0).sum()
viajaron_con_familia = (df["Familiares"] > 0).sum()
print("Viajaron solos:", viajaron_solos)
print("Viajaron con familiares:", viajaron_con_familia)
```
