```python
# 10. Crear la variable Familiares, clasificar grupos,  y mostrar la distribución de cabinas por grupo familiar

import pandas as pd

# Cargar datos (ajusta la ruta si es necesario)
train = pd.read_csv('../data/train.csv')

# Crear variable Familiares
train["Familiares"] = train["SibSp"] + train["Parch"]

# Clasificar grupos familiares
def clasificar_grupo(n):
    if n == 0:
        return "Grupo 1: sin familiares"
    elif 1 <= n <= 3:
        return "Grupo 2: familias pequeñas"
    else:
        return "Grupo 3: familias grandes"

train["Grupo_Familiar"] = train["Familiares"].apply(clasificar_grupo)

# Distribución de cabinas por grupo familiar
distribucion = (
    train.groupby("Grupo_Familiar")["Cabin"]
         .value_counts(dropna=False)
         .rename("Conteo")
         .reset_index()
)

# Mostrar resultado en VSCode (print)
print("\n=== Distribución de Cabinas por Grupo Familiar ===")
print(distribucion)
```
