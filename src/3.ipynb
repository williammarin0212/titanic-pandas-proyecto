```python
# 3. Genere las estad´ısticas b´asicas de cada dataframe y haga comparaciones entre los dataframes

import pandas as pd

# Cargar los datasets
train_df = pd.read_csv('../data/train.csv')
test_df = pd.read_csv('../data/test.csv')

# Estadísticas básicas de variables numéricas
print("=== Estadísticas del dataset de entrenamiento ===")
display(train_df.describe())

print("\n=== Estadísticas del dataset de prueba ===")
display(test_df.describe())

# Comparación resumida
comparison = pd.DataFrame({
    'Train_mean': train_df.describe().loc['mean'],
    'Test_mean': test_df.describe().loc['mean']
})
print("\n=== Comparación de medias entre entrenamiento y prueba ===")
display(comparison)
```
