```python
import pandas as pd
import os

# Ruta correcta para tu estructura de carpetas
train_df = pd.read_csv(os.path.join("..", "data", "train.csv"))
test_df  = pd.read_csv(os.path.join("..", "data", "test.csv"))

print("Datos cargados correctamente.")
print("Filas train:", len(train_df))
print("Filas test:", len(test_df))
```

```python
# 4. Crear nueva variable "Familiares" (SibSp + Parch)

train_df['Familiares'] = train_df['SibSp'] + train_df['Parch']
test_df['Familiares'] = test_df['SibSp'] + test_df['Parch']

display(train_df[['SibSp', 'Parch', 'Familiares']].head())
display(test_df[['SibSp', 'Parch', 'Familiares']].head())

print("\n=== Estadísticas de Familiares (train) ===")
display(train_df['Familiares'].describe())

print("\n=== Estadísticas de Familiares (test) ===")
display(test_df['Familiares'].describe())
```
