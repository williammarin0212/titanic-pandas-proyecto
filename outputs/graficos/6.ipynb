```python
import pandas as pd
import os

# Ruta correcta para tu estructura de carpetas
train_df = pd.read_csv('../../data/train.csv')
test_df = pd.read_csv('../../data/test.csv')

print("Datos cargados correctamente.")
print("Filas train:", len(train_df))
print("Filas test:", len(test_df))
```

```python
# Celda de seguridad: asegurar que Familiares exista SIEMPRE

if 'Familiares' not in train_df.columns:
    train_df['Familiares'] = train_df['SibSp'] + train_df['Parch']
    print("'Familiares' creado en train_df")

if 'Familiares' not in test_df.columns:
    test_df['Familiares'] = test_df['SibSp'] + test_df['Parch']
    print("'Familiares' creado en test_df")

print("Columnas disponibles en train_df:", train_df.columns.tolist())
print("Columnas disponibles en test_df:", test_df.columns.tolist())
```

```python
# 6. Revise las estad´ısticas nuevamente y analice (discuta) si los conjuntos de datos train.csv y test.csv son representativos del conjunto de datos del item anterior

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Asumiendo que train_df y test_df ya están cargados

# Estadísticas básicas
print("=== Estadísticas de Train ===")
display(train_df.describe())

print("\n=== Estadísticas de Test ===")
display(test_df.describe())

# Comparación de medias
comparacion_medias = pd.DataFrame({
    'Train_mean': train_df.mean(numeric_only=True),
    'Test_mean': test_df.mean(numeric_only=True)
})
print("\n=== Comparación de medias ===")
display(comparacion_medias)

# Visualización de variables numéricas principales
numeric_vars = ['Age', 'Fare', 'SibSp', 'Parch', 'Familiares', 'Pclass']

for var in numeric_vars:
    plt.figure(figsize=(8,4))
    sns.kdeplot(train_df[var], label='Train', fill=True)
    sns.kdeplot(test_df[var], label='Test', fill=True)
    plt.title(f'Distribución de {var} en Train vs Test')
    plt.xlabel(var)
    plt.ylabel('Densidad')
    plt.legend()
    plt.show()

# Comparar proporciones de clases (categorical)
for col in ['Pclass', 'Sex', 'Embarked']:
    plt.figure(figsize=(6,3))
    train_df[col].value_counts(normalize=True).plot(kind='bar', alpha=0.5, label='Train')
    test_df[col].value_counts(normalize=True).plot(kind='bar', alpha=0.5, label='Test')
    plt.title(f'Proporciones de {col} en Train vs Test')
    plt.ylabel('Proporción')
    plt.legend()
    plt.show()
```
